# [D - Derangement](https://atcoder.jp/contests/abc072/tasks/arc082_b)

### [解説](https://img.atcoder.jp/arc082/editorial.pdf)
```
p_i != i : o
p_i == i : x

ゴール
P : ooooo

oo --(swap)-> xx
ox --(swap)-> xx
xo --(swap)-> xx

この操作を繰り返せば良い
```

````c++
#include <bits/stdc++.h>
using namespace std;
typedef long long ll;
#define rep(i, n) for (int i = 0; i < (int)(n); i++)
````

````c++
int N; cin >> N;
vector<int> P(N);
rep(i, N) {
    int p; cin >> p;
    P[i] = p-1;
}

P
````

````c++
template <typename T>
void print_vector(vector<T>& vec) {
  cerr << "[ ";
  for (int i = 0; i < vec.size(); i++) {
    if (i < vec.size() - 1) cerr << vec.at(i) << " ";
    else cerr << vec.at(i);
  }
  cerr << " ]" << endl;
}
````

````c++
ll cnt = 0;
bool is_swapped = true;

is_swapped = false;
for (int i = 0; i < N-1; i++) {
    bool a = P[i] == i;
    bool b = P[i+1] == i;
    if (a || b) {
        swap(P[i], P[i+1]);
        cnt++;
        is_swapped = true;
    }
}
print_vector(P);
````
