# ヒープの実装

```python
# !apt install graphviz
# !pip install graphviz
from graphviz import Graph
```

```python
g = Graph(format="png")

data = "412873695"

# ノードを設定
for c in data:
    g.node(c)

# edgeを設定
for i in range(len(data)//2):
    g.edge(data[i], data[2*i+1])
    g.edge(data[i], data[2*i+2])
```
